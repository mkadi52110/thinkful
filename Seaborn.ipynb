```python
# We import the file and prepare it for analysis.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv('/Users/Kadibhai/Downloads/Thinkful Data Files/ibm-hr-analytics-employee-attrition-performance.zip')

# Convert Attrition and BusinessTravel variables from string to numeric

d={'No':0, 'Yes':1}
df['Attrition'] = df['Attrition'].map(d).fillna(df['Attrition'])
d={'Travel_Rarely':1, 'Travel_Frequently':2, 'Non-Travel':0}
df['BusinessTravel'] = df['BusinessTravel'].map(d).fillna(df['BusinessTravel'])
d={'No':0, 'Yes':1}
df['OverTime'] = df['OverTime'].map(d).fillna(df['OverTime'])
pd.set_option('display.max_columns', 200)
df.head()
```

```python
plt.hist(df['Age'])
```

```python
plt.boxplot(df['Age'])
```

```python
sns.set(style="darkgrid")

ax = sns.boxplot(x='Age',y='MonthlyIncome',data=df,palette='pastel')  

plt.title('Monthly Income by Age')

sns.despine(offset=10, trim=True)
ax.set(xlabel='Age', ylabel='Monthly Income')

plt.show()
```

```python
plt.scatter(x=df['Age'],y=df['MonthlyIncome'])
```

```python
g = sns.lmplot(y='MonthlyIncome', 
               x='Age',
               data=df, # Data.
               fit_reg=True, # The regression line also includes a 95% confidence envelope.
               scatter_kws={'alpha':0.4})
g.set_ylabels("MonthlyIncome")
g.set_xlabels("Age")
plt.title('Scatterplot with regression line: Monthly Income by Age')
plt.show()
```

```python
sns.set(style="darkgrid")

ax = sns.boxplot(x='Attrition',y='Age',data=df,palette='pastel')  

plt.title('Attrition by Age')

sns.despine(offset=10, trim=True)
ax.set(xlabel='Attrition', ylabel='Age')

plt.show()
```

```python
plt.plot(df['Attrition'],df['Age'])
```

```python
sns.set(style="darkgrid")

ax = sns.boxplot(x='Attrition',y='Age',hue='Gender',data=df,palette='pastel')  

plt.title('Attrition by Age and Gender')

sns.despine(offset=10, trim=True)
ax.set(xlabel='Attrition', ylabel='Age')

plt.show()
```

```python
g = sns.lmplot(y='OverTime', 
               x='Attrition',
               hue='MaritalStatus',
               data=df, # Data.
               col='Gender',
               fit_reg=True, 
               ci=False,
               scatter_kws={'alpha':0.4}, 
               palette='pastel',
               lowess=False) # Using local weighing to fit a line.
g.set_ylabels("OverTime")
g.set_xlabels("Attrition")
plt.suptitle('Attrition rises with Overtime ')
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
sns.set(style="whitegrid")

g = sns.pointplot(x="Attrition", y="PerformanceRating", hue="JobSatisfaction", data=df,
                   height=6, kind="point", palette="pastel",ci=95,dodge=True,join=True)
plt.title('Pointplot: ')
plt.show()
```

This plot gives us a lot of information. The mean performance has a much larger range for those with attrition and a higher level of performance for those with low job satisfaction. Thus deviation from the mean performance could mean higher probability for attrition.
