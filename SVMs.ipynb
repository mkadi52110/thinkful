```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
raw_data = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/epi_r.csv')
raw_data.head(20)
```

```python
# Count nulls 
null_count = raw_data.isnull().sum()
null_count[null_count>0]
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

clean_data = raw_data.dropna()

X = clean_data.drop('rating',axis=1)
X = clean_data.drop('title',axis=1).sample(frac=0.3, replace=True, random_state=1)

encoder = LabelEncoder()
scaler = StandardScaler()

x_scaled = scaler.fit_transform(X)

Y = clean_data.rating.sample(frac=0.3, replace=True, random_state=1)
y_encoded = encoder.fit_transform(Y)
```

```python
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn import preprocessing
from sklearn import utils

pca = PCA(n_components=30).fit_transform(x_scaled)

svm = SVC(kernel = 'linear')
svm.fit(pca, y_encoded)
```

```python
svm.score(pca, y_encoded)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(svm, pca, y_encoded, cv=5)
```
